# Исследование объявлений о продаже квартир

## Описание проекта

В нашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктов за несколько лет. 

Задача — выполнить предобработку данных и изучить их, чтобы найти интересные особенности и зависимости, которые существуют на рынке недвижимости.

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

## Описание данных

- `airports_nearest` — расстояние до ближайшего аэропорта в метрах (м)
- `balcony` — число балконов

- `floors_total` — всего этажей в доме

- `kitchen_area` — площадь кухни в квадратных метрах (м²)
- `last_price` — цена на момент снятия с публикации
- `living_area` — жилая площадь в квадратных метрах (м²)
- `locality_name` — название населённого пункта
- `open_plan` — свободная планировка (булев тип)
- `parks_around3000` — число парков в радиусе 3 км
- `parks_nearest` — расстояние до ближайшего парка (м)
- `ponds_around3000` — число водоёмов в радиусе 3 км

- `rooms` — число комнат
- `studio` — квартира-студия (булев тип)
- `total_area` — общая площадь квартиры в квадратных метрах (м²)
- `total_images` — число фотографий квартиры в объявлении

**Работа с пропусками**

Выше были обнаружены пропуски в 14 столбцах: ceiling_height, floors_total, living_area, is_apartment, kitchen_area, balcony, locality_name, airports_nearest, cityCenters_nearest, parks_around3000, parks_nearest, ponds_around3000, ponds_nearest, days_exposition.

Пропуски в столбцах is_apartment и balcony заполним значениями False и 0, так как скорее всего пропуски в этих столбцах означают, что это не апартаменты и балконов там нет.

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
data['balcony'] = data['balcony'].fillna(0)
```

Для заполнения пропусков в столбце с высотой потолков, будем использовать медиану. Ожидаемые границы высоты потолков: нижняя - по нормам сан.пин. 2.4м, верхняя - около 6м (высота потолков в пентхаусах или двухуровневых квартирах). Так как по гистограмме видно, что в столбце присутсвуют выбросы, то сперва избавимся от них. 

```python
data.boxplot(column='ceiling_height')
```

Высота потолка в 100м это явная аномалия. Посмотрим на строки где высота потолка выше 7 м

```python
data.query('ceiling_height > 7')
```

Если высоту потолка в 8 м, еще можно представить где то в двухуровневой квартире, то вот 14м, 25м и т.д. это явно неверно записанные значения, которые скорее всего соответсвуют 1.4м или 2.5м соответсвенно. Произведем замену

```python
data['ceiling_height'] = data['ceiling_height'].where(data['ceiling_height'] < 10, data['ceiling_height'] / 10)
data.boxplot(column='ceiling_height', figsize=(3, 5))
```

Теперь нормальные значения расположились в интревале от 2.2 и 3.2, что является адекватными размерами, рассчитаем медиану и заполним ей пропуски.

```python
ceiling_height_median = data['ceiling_height'].median()
ceiling_height_median
```

```python
data['ceiling_height'] = data['ceiling_height'].fillna(ceiling_height_median)
```

Для заполнения пропусков в столбце living_area рассчитаем коэфициент соотношения средней жилой площади (k1) к общей площади. При этом с учетом количества комнат в квартире, так как под жилой площадью как раз и понимается площадь жилых комнат. И затем заполним пропуски в каждом столбце рассчитав его как произведение общей площади на полученный коэффициент.

В столбце с общей площадью, встречаются выбросы, квартира в 800 кв. м. звучит невероятно, посмотрим может быть в данных закралась ошибка и это не 800, а 80кв.м.

```python
data.query('total_area > 200')
```

Ошибки нет, эти квартиры и стоят соответсвенно, но таких квартир очень мало и от них лучше избавится. Посмотрим в каком интервале расположились нормальные значения.

```python
data.boxplot(column='total_area')
plt.ylim(0, 200)
```

```python
data['total_area'].hist(range=(12, 150), bins=100)
```

Нормальные значения расположились в интервале от 22 до 130 кв.м. Удалим выбросы.

```python
data = data.query('22 <= total_area <= 130').reset_index(drop=True)
```

```python
data['rooms'].value_counts() # посмотрим скольки комнатные квартиры остались
```

1, 2, 3 и 4 комнатные квартиры составляют около 95% всех оставшихся значений, остальные встречаются крайне редко, поэтому избавимся и от них, и перейдем к расчету коэффициента.

```python
data = data.query('1 <= rooms <= 4').reset_index(drop=True)
```

```python
k1 = data.groupby('rooms')['living_area'].mean() / data.groupby('rooms')['total_area'].mean() 
k1 # создан Series со значениями к1 для каждого количества камнат
```

```python
def living_area_fillna(row): # функция заполняющая пропуски в столбце с жилой площадью, с учетом количества комнат
    if pd.isna(row['living_area']):
        return row['total_area'] * k1[row['rooms']]
    else:
        return row['living_area']
```

```python
data['living_area'] = data.apply(living_area_fillna, axis=1)
```

Применим аналогичные рассуждения для заполнения пропусков в столбце площадь кухни.

```python
k2 = data.groupby('rooms')['kitchen_area'].mean() / data.groupby('rooms')['total_area'].mean() 
k2 # создан Series со значениями отношения площади кухни к общей площади для каждого количества камнат
```

```python
def kitchen_area_fillna(row): # функция заполняющая пропуски в столбце с площадью кухни, с учетом количества комнат
    if pd.isna(row['kitchen_area']):
        return row['total_area'] * k2[row['rooms']]
    else:
        return row['kitchen_area']
```

```python
data['kitchen_area'] = data.apply(kitchen_area_fillna, axis=1)
```

Посмотрим какие названия локаций есть в датасете. Но перед этим удалим все пропуски в этом столбце.

```python
data = data.query('not locality_name.isna()')
```

```python
sorted(data['locality_name'].unique())
```

Неявные дубликаты могут скрываться: в названиях со словами "поселок" и "посёлок", в названиях со словами "поселок городского типа" и "городской поселок" и т.д. Так же возможна ситуация что в датасете, например будут содержаться "деревня Батово" и "Батово". Поэтому на мой взгляд лучшим решением будет избавится от типа поселения вовсе, оставив только его название.

```python
def create_locality_name(name): # функция которая отбросит все типы поселения, оставив читсые названия
    if 'посёлок при железнодорожной станции' in name:
        name = ' '.join(name.split()[4:])
    if (('поселок городского типа' in name) or ('посёлок городского типа' in name) or 
        ('садоводческое некоммерческое товарищество' in name)):
        name = ' '.join(name.split()[3:])
    if (('городской поселок' in name) or ('городской посёлок' in name) or 
        ('коттеджный поселок' in name) or ('коттеджный посёлок' in name) or 
        ('садовое товарищество' in name) or ('поселок станции' in name) or
        ('посёлок станции' in name)):
        name = ' '.join(name.split()[2:])
    if ('деревня' in name) or ('поселок' in name) or ('посёлок' in name) or ('село' in name):
        name = ' '.join(name.split()[1:])
    return name
```

```python
data['locality_name'] = data['locality_name'].apply(create_locality_name) # применим функцию к столбцу с названиями
sorted(data['locality_name'].unique()) # посмотрим что получилось
```

1. Предобработка данных. На данном этапе были заполнены пропуски, обработаны аномальные значения а так же преобразованы типы данных для некоторых столбцов.
2. Исследовательский анализ данных. На данном этапе были изучен ряд параметров, влияющих на стоимость недвижимости, построены гистограммы и выявлены зависимости.
3. Общий вывод:
    1. Наибольшее влияние на цену оказывает именно общая площадь. Чем больше общая площадь, тем больше цена.
    2. Медианная цена недвижимости 4 - 5 млн.
    2. В будние дни в среднем размещаются больше объявлений чем в выходные, при этом стоимость квартир размещенных в будни выше чем в выходные. Можно предположить, что в будние дни объявления размещают агенты, которые включают в стоимость недвижимости свой процент.
    3. В 2014 году был пик цен на недвижимость - средняя стоимость более 8 млн, скорее всего связанный с резким скачком роста доллара. После набдюдается спад вплоть до 2018 года - минимум около 5 млн рублей. В 2019 - рост.
    4. Чаще самые дешевые первые этажи, затем последние, затем любые другие.
    5. После 5 км от центра города, цены стремительно падают, логично предположить, что на пятом километре заканчивается центр города.
